```python
import pandas as pd
from googletrans import Translator

translator = Translator()

```

```python
df = pd.read_csv('nepali_poems.csv')
```

```python
df
```

```python
def translate_text(text):
    return translator.translate(text, src='ne', dest='en').text
    

```

```python
df['Translated_Text'] = df['Content'].apply(translate_text)

df[['Title', 'Translated_Text']].to_csv('translated_poems.csv', index=False)
```
